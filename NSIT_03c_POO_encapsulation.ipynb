## Attributs privés

Cela peut être résolu en rendant les attributs <b>privés</b>.  
En Python, par défaut, tous les attributs sont <b>publiques</b>: c'est-à-dire visibles et modifiables par tout autre objet ou programme extérieur.  
Dans certains cas, cela peut devenir dangeureux et provoquer des incohérences (Comme ci-dessus).

S'il on veut les rendre <b>privés</b> ,on peut placer un <b>double tiret bas</b> (ou <i>underscrore</i> en anglais) devant le nom de la variable.  
Par exemple: '\__nb_passagers' serait la version privée de l'attribut 'nb_passagers'.  
De la même façon, on peut rendre des méthodes privées: Dans le code ci-dessous, on ajoute la méthode '\__places_libres()'. Elles ne peut être appelée que par l'objet lui-même.    
Voila à quoi ressemble alors notre classe avec ce filet de sécurité.

```python
class Rame_metro2:
    def __init__(self, reference_rame, nb_passagers = 0, nb_passagers_max = 500):
        self.__reference_rame = reference_rame
        self.__nb_passagers = nb_passagers
        self.__nb_passagers_max = nb_passagers_max
    def depose(self, en_moins:int)->int:
        """ Les personnes 'en_moins' descendent du métro
            Pre-condition: 'en_moins' ne devrait pas être supérieur au nombre actuel de passagers.
            La fonction renvoie le nombre actuel de passagers.
        """
        self.__nb_passagers -= en_moins
        if self.__nb_passagers < 0:
            self.__nb_passagers = 0
        print("il reste", self.__places_libres(),"places libres")
        return self.__nb_passagers
    
    def charge(self, en_plus:int)->int:
        """ Les personnes 'en plus' tentent de rentrer dans le métro
            Post-condition: 'nb_passagers' ne pourra pas dépasser 'self.nb_passagers_max'
            La fonction renvoie le nombre de personnes restées sur le quai
        """
        self.__nb_passagers += en_plus
        if self.__nb_passagers > self.__nb_passagers_max : 
            reste_sur_le_quai = self.__nb_passagers - self.__nb_passagers_max
            self.__nb_passagers = self.__nb_passagers_max
            return reste_sur_le_quai  # Ces gens n'ont pas pu rentrer.
        return 0 # Tout le monde est rentré dans le métro. Plus personne sur le quai
    
    def __places_libres(self)->int:
        """ Renvoie la différence entre le nombre de place max et les places prises.
        """
        return self.__nb_passagers_max - self.__nb_passagers
        
# Instanciation d'une rame
rame_3 = Rame_metro2("MF88_3", 200, 800)
```

La  méthode 'depose()' renvoie ici 250.  
En effet, la rame est instanciée avec 200 passagers auxquels nous avons précédemment ajouté 100 personnes, et nous venons d'en déposer 50.  
200 + 100 - 50 =  250  
De plus nous voyons que cette méthode fait appel à '\__places_libres()' et elle affiche qu'il reste 550 places lilbres.  
800 - 250 = 550

Mais, si on souhaite lancer la méthode '\__places_libres()' depuis l'extérieur de l'objet, il ne va pas se montrer coopératif:

```python
rame_3.__places_libres()
```

Si maintenant on tente d'interroger la valeur de l'attribut \__nb_passagers, l'interpréteur se comporte comme si elle n'exitait pas!

```python
rame_3.__nb_passagers
```

'depose()' et 'charge()' sont deux méthodes qui, en interne, ont accès aux attributs, et on le droit de les modifier: Ce type de méthode porte le nom de <b>mutateur</b> (ou <b>setters</b> en anglais).  

Si l'on souhaite rendre ces attibuts visibles, nous devons créer des méthodes spécifiques qui en interne ont le droit de voir ces valeurs, pour les afficher: On appelle ces méthodes des <b>accesseurs</b> (ou <b>getters</b> en anglais)

## Encapsulation

Voila , finalement, à quoi ressemble un code un peu plus sécurisé de notre classe:

```python
class Rame_ligne_7bis:
    def __init__(self, reference_rame, nb_passagers = 0, nb_passagers_max = 500):
        self.__reference_rame = reference_rame
        self.__nb_passagers = nb_passagers
        self.__nb_passagers_max = nb_passagers_max
    def depose(self, en_moins:int)->int:
        """ Les personnes 'en_moins' descendent du métro
            Pre-condition: 'en_moins' ne devrait pas être supérieur au nombre actuel de passagers.
            La fonction renvoie le nombre actuel de passagers.
        """
        self.__nb_passagers -= en_moins
        if self.__nb_passagers < 0:
            self.__nb_passagers = 0
        return self.__nb_passagers
    
    def charge(self, en_plus:int)->int:
        """ Les personnes 'en plus' tentent de rentrer dans le métro
            Post-condition: 'nb_passagers' ne pourra pas dépasser 'self.nb_passagers_max'
            La fonction renvoie le nombre de personnes restées sur le quai
        """
        self.__nb_passagers += en_plus
        if self.__nb_passagers > self.__nb_passagers_max : 
            reste_sur_le_quai = self.__nb_passagers - self.__nb_passagers_max
            self.__nb_passagers = self.__nb_passagers_max
            return reste_sur_le_quai  # Ces gens n'ont pas pu rentrer.
        return 0 # Tout le monde est rentré dans le métro. Plus personne sur le quai
    
    def get_nb_passagers(self)->int:
        """ Accesseur de l'attribut privé __nb_passagers
        """
        return self.__nb_passagers
    
    def get_nb_passagers_max(self)->int:
        """ Accesseur de l'attribut privé __nb_passagers_max
        """
        return self.__nb_passagers_max
        
# Instanciation d'une rame
rame_3 = Rame_ligne_7bis("MF88_3", 200, 800)

# On regarde ce qui se passe si on tente de faire descendre 210 personnes alors qu'il n'y en a que 200 dedans
rame_3.depose(210)
# Normalement, nous ne devrions pas obtenir un nombre négatif de passagers
print("On tente de déposer 210 personnes. Il en reste",rame_3.get_nb_passagers())
# On faire remonter 500 personnes
sur_le_premier_quai = rame_3.charge(500)
print("500 personnes montent. Il y en a maintenant",rame_3.get_nb_passagers(),"dans le métro")
# Puis 600 personnes de plus se présentent sur le quai
sur_le_second_quai =rame_3.charge(600)
print("600 de plus. Il y en a maintenant",rame_3.get_nb_passagers(),
      "dans le métro, et", sur_le_second_quai,"sont restés sur le quai")
```

Voilà!  
Les <b>accesseurs</b> me permettent de voir certains attributs, mais je ne peux plus les modifier sans passer par les <b>mutateurs</b>.

Cette protection illustre le principe d'<b>encapsulation</b>:  
Nos données sont bien empactées dans une capsule protectrice.

Dans notre cas, je peux être certain que le nombre de passagers sera maîtrisé, et, toujours compris en 0 et '\__nb_passagers_max'.
